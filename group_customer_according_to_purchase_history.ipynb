```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 
from sklearn.cluster import KMeans 
%matplotlib inline
```

```python
customer_data = pd.read_csv("Segmentation_dataset.csv")
```

```python
fig,ax=plt.subplots(2,figsize=(9,18))
sns.scatterplot(ax=ax[0],data=customer_data,x='Age',y='Spending Score (1-100)',hue='Gender')
sns.scatterplot(ax=ax[1],data=customer_data,x='Annual Income (k$)',y='Spending Score (1-100)',hue='Gender')
plt.show()
```

```python
sns.pairplot(customer_data,hue='Gender')
```

```python
plt.figure(1 , figsize = (16 , 8))
n = 0 
for cols in ['Age' , 'Annual Income (k$)' , 'Spending Score (1-100)']:
    n += 1 
    plt.subplot(1 , 3 , n)
    plt.subplots_adjust(hspace = 0.5 , wspace = 0.5)
    sns.violinplot(x = cols , y = 'Gender' , data = customer_data , palette = 'vlag')
    sns.swarmplot(x = cols , y = 'Gender' , data = customer_data,alpha=0.6, size=3)
    plt.ylabel('Gender' if n == 1 else '')
    plt.title('Boxplots & Swarmplots' if n == 2 else '')
plt.show()
```

```python
sse=[]
for k in range(1,20):
    kmk=KMeans(n_clusters=k)
    kmk.fit(customer_data[['Annual Income (k$)','Spending Score (1-100)']])
    sse.append(kmk.inertia_)
```

```python
sns.lineplot(data=sse)
sns.scatterplot(data=sse)
plt.grid(True)
plt.xlabel('Number of Clusters')
plt.ylabel('Sum of Squared Error')
plt.title('Elbow Chart')
plt.show()
```

```python
A = customer_data.drop(columns=['CustomerID','Gender','Age'],axis=1).values
kmk=KMeans(n_clusters=5,init='k-means++',random_state=42)
B = kmk.fit_predict(A)
```

```python
kmk.cluster_centers_
```

```python
plt.figure(figsize=(16,10))
plt.scatter(A[B==0,0], A[B==0,1] , s=50 , c='red' , label='cluster 1')
plt.scatter(A[B==1,0], A[B==1,1] , s=50 , c='blue' , label='cluster 2')
plt.scatter(A[B==2,0], A[B==2,1] , s=50 , c='yellow' , label='cluster 3')
plt.scatter(A[B==3,0], A[B==3,1] , s=50 , c='green' , label='cluster 4')
plt.scatter(A[B==4,0], A[B==4,1] , s=50 , c='orange' , label='cluster 5')
plt.scatter(kmk.cluster_centers_[:,0],kmk.cluster_centers_[:,1], s=200 , c='black' , label='cluster centre')
plt.title('customer Groups')
plt.xlabel("Annual income")
plt.ylabel('Spending Score')
plt.show()
```
